```python
# 讀取csv檔
url = 'https://drive.google.com/uc?id=1VKFAoB_bPhUilE79LSnCHyJaDNps2NfF&export=download'
df = pd.read_csv(url, encoding='utf-8')
```

```python
# ans4
avg_temp = df["氣溫(℃)"].mean()
max_humi = df['相對溼度(%)'].max()
min_pres = df['當日氣壓(hPa)'].min()
tal_rain = df['降水量(mm)'].sum()

print("平均溫度:",avg_temp,"℃")
print("最高溼度:",max_humi,"%")
print("最低氣壓:",min_pres,"hPa")
print("總雨量:",tal_rain,"mm")
```

```python
# ans5
plt.plot(df["觀測日期(day)"], df["氣溫(℃)"], "b-",label = "氣溫")
plt.title("6月氣溫變化",fontproperties=myfont)
plt.xlabel("日期",fontproperties=myfont)
plt.ylabel("氣溫(℃)",fontproperties=myfont)
plt.legend(loc = 'best',prop=myfont)
plt.show()
```

# 額外功能

```python
#計算三日以內氣溫平均

#計算方式
days = 3 # 方便更改
df["三日均溫"] = df["氣溫(℃)"].rolling(days).mean()

# 圖表繪製
plt.plot(df["觀測日期(day)"], df["氣溫(℃)"], "b-",label = "當日氣溫")
plt.plot(df["觀測日期(day)"], df["三日均溫"], "r-.",label = "三日均溫")
plt.title("6月氣溫變化",fontproperties=myfont)
plt.xlabel("日期",fontproperties=myfont)
plt.ylabel("氣溫(℃)",fontproperties=myfont)
plt.legend(loc = 'best',prop=myfont)

# 結果列印
plt.show()
print(df.loc[:, ["三日均溫","氣溫(℃)"]])
```
